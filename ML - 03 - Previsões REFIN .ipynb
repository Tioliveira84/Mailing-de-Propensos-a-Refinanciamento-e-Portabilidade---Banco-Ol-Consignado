***
<font size="4"><a id="info">    03.1 - codificação one-hot</a></font>
***

<div class="alert alert-block alert-success">
<b>Descrição: Realização do Dummie  </b> </b>
</div>

```python
arOneHotEncoding = ['STATUS','CANAL','GRPCONV']
dfArquivoTreinamento = pd.get_dummies(df_data,columns=arOneHotEncoding)
dfArquivoTreinamento.head()
```

***
<font size="6"><a id="info">03.2 - Normalização</a></font>
***

<div class="alert alert-block alert-success">
<b>Descrição: </b> </b>
</div>

```python
arNormalize = ['TAXA','QUANT_PRESTACAO','VALOR_FINANCIADO','VALOR_PRESTACAO','DURACAO_CONTRATO']
scaler = MinMaxScaler(feature_range=(0,1))

for eachCategory in arNormalize:
    dfArquivoTreinamento[eachCategory] = scaler.fit_transform(dfArquivoTreinamento[eachCategory].values.reshape(-1, 1))
```

```python
dfArquivoTreinamento.head()
```

***
<font size="6"><a id="info">03.3 - Remover Colunas não necessárias</a></font>
***

<div class="alert alert-block alert-success">
<b>Descrição: </b> </b>
</div>

```python
arColunsToRemove = ['NRO_CONTRATO','CNPJ_CPF']
dfArquivoTreinamento.drop(columns=arColunsToRemove, inplace=True,axis=1)
```

```python
dfArquivoTreinamento.shape
  
```

```python
arquivo = 'C:/Users/bsi90720/Desktop/Modelo de Proprensão Refin Port - Preparação/REFIN/colunas.csv'

dfcolunas = pd.read_csv(arquivo)
dffinal=pd.DataFrame()
```

```python
dfcolunas['colunas'].values
```

```python
arZeros = np.zeros(dfArquivoTreinamento.shape[0])
for col in dfcolunas['colunas'].values:
    try:
        dffinal[col] = dfArquivoTreinamento[col]
    except:
        print('Coluna "{}" não encontrada. Preencher com zeros'.format(col))
        dffinal[col] = arZeros
dffinal.head()
```

```python
dffinal.shape
```

***
<font size="6"><a id="keras_1">10.1 - Carregar Modelo Treinado</a></font>
***
[VOLTAR](#sumario)

```python
filenameH5 = 'C:/Users/bsi90720/Desktop/Modelo de Proprensão Refin Port - Preparação/REFIN/RefinPredictionModel-Keras.h5'
filenameJSON = 'C:/Users/bsi90720/Desktop/Modelo de Proprensão Refin Port - Preparação/REFIN/RefinPredictionModel-Keras.json'

json_file = io.open(filenameJSON, 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)

loaded_model.load_weights(filenameH5)
 
loaded_model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
```

***
<font size="6"><a id="keras_2">10.2 - Gerar Previsões</a></font>
***
[VOLTAR](#sumario)

```python
ypredKeras = loaded_model.predict(dffinal.values)
```

```python
yClassesKeras = loaded_model.predict_classes(dffinal.values)
```

```python
dfPrevisoes = pd.DataFrame(data=ypredKeras.reshape(len(ypredKeras)),columns=['PrevisoesRL'])
dfClasses =  pd.DataFrame(data=yClassesKeras.reshape(len(yClassesKeras)),columns=['ClassesRL'])
print(dfPrevisoes.shape)
print(dfClasses.shape)
```

```python
display(dfPrevisoes.head())
display(dfClasses.head())
```

<font size="6"><a id="keras_2">11 - Gerar Saída de Previsão</a></font>

```python
sNomeArquivo = 'C:/Users/bsi90720/Desktop/Modelo de Proprensão Refin Port - Preparação/REFIN/Resultado_propensos_refin.csv'

dfSaida = pd.concat([df_data[['NRO_CONTRATO','CNPJ_CPF']],dfClasses],axis=1)
dfSaida = pd.concat([dfSaida,dfPrevisoes],axis=1)

dfSaida.rename(columns={'NRO_CONTRATO': 'NUMCTT', 'CNPJ_CPF': 'NUMCPF', 'ClassesRL': 'FLAG_REFIN', 'PrevisoesRL': 'LP_FLAG_REFIN'}, inplace=True)

#dfSaida.to_csv(path_or_buf=sNomeArquivo,index=False,sep=",",decimal='.')
```

```python
dfSaida.head(10000)
```

```python
dfSaida.sort_values(['FLAG_REFIN'], ascending=True).groupby(['FLAG_REFIN']).count()[['NUMCTT']]
```

```python
def reclassificar(LP_FLAG_REFIN):
  if LP_FLAG_REFIN < 0.02:
      return 1
  else:
       return 0
```

```python
dfSaida['FLAG_REFIN'] = np.array([reclassificar(LP_FLAG_REFIN) for LP_FLAG_REFIN in dfSaida['LP_FLAG_REFIN']])
```

```python
dfSaida.sort_values(['FLAG_REFIN'], ascending=True).groupby(['FLAG_REFIN']).count()[['NUMCTT']]
```
